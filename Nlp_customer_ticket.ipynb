```python
with open("config.cfg", "r") as f:
    config = f.read()

new_config = config.replace('vectors = "en_core_web_lg"', 'vectors = "en_core_web_md"')  # ou toute autre ligne à modifier

# Réécrire le fichier
with open("config.cfg", "w") as f:
    f.write(new_config)
```

```python
import mlflow
mlflow.set_tracking_uri("/content/drive/MyDrive/Nlp_Customer_ticket/mlruns") #("file:///content/mlruns")
mlflow.set_experiment("Custumer_ticket_queue_gathered")
```

```python

post_fix=datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
cfg = load_config("config.cfg")
for i in range(2):
  with mlflow.start_run(run_name="en_core_web_md"+post_fix):
      mlflow.log_artifact("config.cfg", "data" )
      mlflow.log_artifact("train_queue_gathering.spacy", "data")
      mlflow.log_artifact("valid_queue_gathering.spacy", "data")

      mlflow.log_param("language", cfg["nlp"]["lang"])
      mlflow.log_param("pipeline", ",".join(cfg["nlp"]["pipeline"]))
      mlflow.log_param("batch_size", cfg["nlp"]["batch_size"])
      mlflow.log_param("dropout", cfg["training"]["dropout"])
      mlflow.log_param("max_steps", cfg["training"]["max_steps"])
      mlflow.log_param("max_epochs", cfg["training"]["max_epochs"])

      !python -m spacy train config.cfg --output ./output --paths.train "/content/drive/MyDrive/Nlp_Customer_ticket/Data/Spacy_format/train_queue_gathering.spacy" --paths.dev "/content/drive/MyDrive/Nlp_Customer_ticket/Data/Spacy_format/valid_queue_gathering.spacy" --gpu-id 0

      mlflow.log_metric
      mlflow.log_artifacts("output")
```
